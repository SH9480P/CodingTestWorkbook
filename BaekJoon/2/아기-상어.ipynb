```python
from collections import deque

def find_shark():
    for i in range(N):
        for j in range(N):
            if board[i][j] == 9:
                return i, j

def isGridIn(x, y):
    return 0 <= x < N and 0 <= y < N

N = int(input())
board = [list(map(int, input().split())) for _ in range(N)]
visited = [[False]*N for _ in range(N)]
ULRD = [(-1, 0), (0, -1), (0, 1), (1, 0)]
r, c = find_shark()
size = 2
ate = 0
seconds = 0

cur_queue = deque([(r, c)])
visited[r][c] = True
board[r][c] = 0
dist = 0
while True:
    cand = []
    next_queue = deque()
    while cur_queue:
        x, y = cur_queue.popleft()
        if 0 < board[x][y] < size:
            cand.append((x, y))
        for dx, dy in ULRD:
            nx, ny = x+dx, y+dy
            if isGridIn(nx, ny) and not visited[nx][ny] and board[nx][ny] <= size:
                next_queue.append((nx, ny))
                visited[nx][ny] = True
    if cand:
        feed_x, feed_y = min(cand)
        ate += 1
        if ate == size:
            size += 1
            ate = 0
        r, c = feed_x, feed_y
        board[r][c] = 0
        cur_queue.append((r, c))
        visited = [[False] * N for _ in range(N)]
        visited[r][c] = True
        seconds += dist
        dist = 0
        continue
    if next_queue:
        cur_queue = next_queue
    else:
        break
    dist += 1
print(seconds)
```
