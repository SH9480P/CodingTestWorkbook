# 주사위 굴리기

소요시간: NaN

작성일시: 24.03.20 01:04:42

출처: https://www.acmicpc.net/problem/14499

### 접근 방법
주사위를 굴렸을 때 전후좌우상하를 파악하는 것이 관건이다.  
1차원 배열에 주사위 인덱스를 넣어놓고, 상하좌우로 굴렸을 때 어떻게 변화하는지 규칙을 찾으면 되는 문제.

### 시간 복잡도 분석
O(K)

### 새로 알게 된 것
.

### 주의할 점
.

### 기타 코멘트
.


```python
import sys

input = sys.stdin.readline

def isGridIn(r, c):
    return 0 <= r < N and 0 <= c < M

def left():
    moved = direction[:]
    moved[2] = direction[0]
    moved[0] = direction[1]
    moved[5] = direction[2]
    moved[1] = direction[5]
    return moved

def right():
    moved = direction[:]
    moved[0] = direction[2]
    moved[1] = direction[0]
    moved[2] = direction[5]
    moved[5] = direction[1]
    return moved

def up():
    moved = direction[:]
    moved[4] = direction[0]
    moved[0] = direction[3]
    moved[5] = direction[4]
    moved[3] = direction[5]
    return moved

def down():
    moved = direction[:]
    moved[0] = direction[4]
    moved[3] = direction[0]
    moved[4] = direction[5]
    moved[5] = direction[3]
    return moved

N, M, x, y, K = map(int, input().split())
board = [list(map(int, input().rstrip().split())) for _ in range(N)]
direction = list(range(6))
dice = [0] * 6
RLUD = [(0, 1), (0, -1), (-1, 0), (1, 0)]
move = [right, left, up, down]
for d in map(int, input().rstrip().split()):
    dx, dy = RLUD[d-1]
    nx, ny = x + dx, y + dy
    if isGridIn(nx, ny):
        x, y = nx, ny
        direction = move[d-1]()
        if board[x][y] == 0:
            board[x][y] = dice[direction[0]]
        else:
            dice[direction[0]] = board[x][y]
            board[x][y] = 0
        print(dice[direction[-1]])
```
