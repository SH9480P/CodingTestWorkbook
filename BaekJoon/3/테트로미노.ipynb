### 접근 방법
1st
- 테트로미노 4종이 가질 수 있는 모든 형태를 2차원 배열로 저장하여 종이의 모든 칸에 대 보았다.
- 시간초과

2nd
- 1st에서 서로 다른 형태의 테트로미노가 갖는 값을 계산하는데 위치가 비슷하면 여러번 반복해서 합산해야 해서 비효율적이라는 생각은 했으나, 이를 어떻게 해결해야 할 지 감을 잡지 못했다.
- dfs로 4칸을 가면 테트로미노를 만들 수 있다...

### 시간 복잡도 분석
O(NM)

```python
import sys

input = sys.stdin.readline

def rotate(arr):
    n = len(arr)
    m = len(arr[0])
    rotated = [[True] * n for _ in range(m)]
    for i in range(n):
        for j in range(m):
            rotated[j][n-1-i] = arr[i][j]
    return rotated

def flip(arr):
    n = len(arr)
    m = len(arr[0])
    flipped = [[True] * m for _ in range(n)]
    for i in range(n):
        for j in range(m):
            flipped[n-1-i][j] = arr[i][j]
    return flipped

def first():
    arr = [[True] * 4]
    rotated = [[True] for _ in range(4)]
    return [arr, rotated]

def second():
    arr = [
        [True, True],
        [True, True]
    ]
    return [arr]

def third():
    result = []
    arr = [
        [True, False],
        [True, False],
        [True, True]
    ]
    flipped = flip(arr)
    for _ in range(4):
        result.append(rotate(arr))
        result.append(rotate(flipped))
    return result

def fourth():
    arr = [
        [True, False],
        [True, True],
        [False, True]
    ]
    flipped = flip(arr)
    return [arr, flipped, rotate(arr), rotate(flipped)]

def fifth():
    arr = [
        [True, True, True],
        [False, True, False]
    ]
    return [rotate(arr) for _ in range(4)]

def isGridIn(r, c):
    return 0 <= r < N and 0 <= c < M

N, M = map(int, input().rstrip().split())
board = [list(map(int, input().rstrip().split())) for _ in range(N)]
tetro = []
tetro.extend(first())
tetro.extend(second())
tetro.extend(third())
tetro.extend(fourth())
tetro.extend(fifth())

result = 0
for t in tetro:
    R = len(t)
    C = len(t[0])
    for i in range(N):
        for j in range(M):
            curVal = 0
            for r in range(R):
                for c in range(C):
                    x, y = i + r, j + c
                    if isGridIn(x, y):
                        if t[r][c]:
                            curVal += board[x][y]
                    else:
                        break
            result = max(result, curVal)

print(result)
```
