## 미로 탈출

출처: https://school.programmers.co.kr/learn/courses/30/lessons/159993

가로 길이 = N
세로 길이 = M
시간복잡도 = O(N*M)

#### 기타 코멘트
- 벽이 있는 칸을 체크하여 탐색하지 못하게 해야 하는 사실을 간과하였음
- isGridIn에 벽이 있는 칸을 체크하는 기능을 추가하면 더 깔끔할 듯

```python
from collections import deque

UDLR = [(-1, 0), (1, 0), (0, -1), (0, 1)]
visited = []

def isGridIn(maps, x, y):
    rowLen = len(maps)
    colLen = len(maps[0])
    return 0 <= x < rowLen and 0 <= y < colLen


def bfs(maps, startPos, endPos):
    q = deque()
    sx, sy = startPos
    q.append((sx, sy, 0))
    visited[sx][sy] = True
    while q:
        x, y, c = q.popleft()
        if (x, y) == endPos:
            return c
        for dx, dy in UDLR:
            nx, ny = x + dx, y + dy
            if isGridIn(maps, nx, ny) and maps[nx][ny] != 'X' and not visited[nx][ny]:
                q.append((nx, ny, c + 1))
                visited[nx][ny] = True
    return -1

def solution(maps):
    global visited
    rowLen = len(maps)
    colLen = len(maps[0])
    visited = [[False] * colLen for _ in range(rowLen)]

    startPos = None
    leverPos = None
    exitPos = None
    for i in range(rowLen):
        for j in range(colLen):
            if maps[i][j] == 'S':
                startPos = (i, j)
            elif maps[i][j] == 'L':
                leverPos = (i, j)
            elif maps[i][j] == 'E':
                exitPos = (i, j)
            if startPos and leverPos and exitPos:
                break

    s2l = bfs(maps, startPos, leverPos)
    if s2l == -1:
        return -1
    visited = [[False] * colLen for _ in range(rowLen)]
    l2e = bfs(maps, leverPos, exitPos)
    if l2e == -1:
        return -1
    return s2l + l2e
```
